🎵 Spotify & YouTube 데이터 컬럼 설명

Track
스포티파이 플랫폼에서 볼 수 있는 곡명

Artist
아티스트의 이름

Url_spotify
아티스트(또는 곡)의 Spotify URL

Album
해당 곡이 스포티파이에 수록된 앨범명

Album_type
해당 곡이 싱글로 발매되었는지, 앨범 수록곡인지 여부

Uri
Spotify API를 통해 노래를 찾는 데 사용하는 고유 링크

🎼 음악적 특성 (Audio Features)

Danceability
템포, 리듬 안정성, 박자 강도 등을 종합하여
트랙이 춤에 얼마나 적합한지를 나타내는 값 (0.0 ~ 1.0)

Energy
강도와 활동성을 나타내는 지표 (0.0 ~ 1.0)
값이 높을수록 빠르고 강렬한 느낌

Key
곡의 조성 (정수값)

0 = C

1 = C♯/D♭

2 = D

...

감지되지 않은 경우 = -1

Loudness
전체 평균 음량 (단위: dB)
일반적으로 -60 ~ 0 dB 범위

Speechiness
트랙 내 말(음성)의 비중 (0.0 ~ 1.0)

0.66 이상 → 대부분 말

0.33 ~ 0.66 → 음악 + 말

0.33 미만 → 대부분 음악

Acousticness
어쿠스틱 곡일 가능성 (0.0 ~ 1.0)
1.0에 가까울수록 어쿠스틱일 확률이 높음

Instrumentalness
보컬이 없을 가능성 (0.0 ~ 1.0)
0.5 이상이면 기악곡일 확률이 높음

Liveness
라이브 공연 녹음 여부 예측 (0.0 ~ 1.0)
0.8 이상이면 라이브일 가능성 높음

Valence
곡이 전달하는 긍정적 감정의 정도 (0.0 ~ 1.0)
높을수록 밝고 긍정적인 느낌

Tempo
곡의 전체 추정 템포 (BPM)

Duration_ms
곡의 길이 (밀리초 단위)

📊 Spotify 성과 지표

스트리밍 (Streams)
스포티파이에서의 총 스트리밍 횟수

📺 YouTube 관련 정보

Url_youtube
해당 곡과 연결된 유튜브 영상 URL

Title
유튜브 영상 제목

Channel
영상을 게시한 채널명

Views
조회수

Likes
좋아요 수

Comments
댓글 수

Description
유튜브 영상 설명

Licensed
라이선스 콘텐츠 여부 (콘텐츠 파트너가 권리를 주장한 영상인지 여부)

official_video
해당 영상이 공식 곡(Official Video)인지 여부 (Boolean)

## Task4 EDA

데이터 셋에서 필요없은 컬럼들을 삭제하겠음!  
-> Unnamed(의미 없는 인덱스임), Url_spotify,Uri, Url_youtube(uri도 의미 없으니 삭제), Description(유튜브 영상 설명글이므로 삭제), Licensed (라이선스), Title, Channel(아티스트, 곡제목과 동일하기 때문에 삭제)

```python
#컬럼 이름 출력
print(list(data.columns))
```

```python
# 삭제할 컬럼 
cols_to_drop = ['Unnamed: 0', 'Url_spotify', 'Uri', 'Url_youtube', 'Description', 'Licensed','Title','Channel']

# 컬럼을 삭제한 데이터를 music에 저장
music = data.drop(columns=cols_to_drop)

music.head()
```

```python
#결측치 처리
music.isnull().sum()
```

```python
#stream을 기준으로 분석을 해야하기 때문에 stream이 없는 행들은 삭제(행이 충분하기 때문에 삭제해도 괜찮을 듯)
music = music.dropna(subset=['Stream'])

music.isnull().sum()
```

```python
#Views, Likes, Comments는 아직 어떻게 처리해야할지 모르겠음 -> 일단 두고
#official_video(공식 영상 여부)부터 처리해보자!
music['official_video'].unique()
```

```python
#결측치 값을 nan으로 설정 + boolean타입에서 문자열 타입으로 변경
music['official_video'] = music['official_video'].fillna('nan')
music['official_video'] = music['official_video'].astype(str)
```

```python
print("결측치 개수:", music['official_video'].isnull().sum())
print("데이터 타입:", music['official_video'].dtype)
```

일단 데이터 처리는 이정도로 해두자!

```python
#스트리밍 수가 가장 높은 음악 10개 뽑아보기
import matplotlib.pyplot as plt

# 1️⃣ Stream 기준 상위 10개 추출
top10 = music.nlargest(10, 'Stream')

# 2️⃣ 그래프 생성 
plt.figure()
plt.bar(top10['Track'], top10['Stream'], color='purple')

# 3️⃣ 축 라벨 및 제목
plt.xlabel('Track')
plt.ylabel('Stream')
plt.title('Top 10 Tracks by Stream')

# 4️⃣ x축 글자 겹침 방지
plt.xticks(rotation=90)

plt.show()
```

```python
#spotify 스트림 수와 유튜뷰 view의 관계 시각화
plt.figure()
plt.scatter(music['Views'], music['Stream'], color='purple')
plt.xlabel('Views')
plt.ylabel('Stream')
plt.title('Scatter Plot of Views vs Stream')
plt.grid()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cols = [
    'Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness',
    'Acousticness', 'Instrumentalness', 'Liveness', 'Valence',
    'Tempo', 'Duration_ms', 'Stream'
]

# 숫자형 + 결측치 제거
corr_matrix = music[cols].astype(float).dropna().corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,
            annot=True,
            fmt=".2f",
            cmap="Blues",
            square=True)

for i in range(corr_matrix.shape[1]):
    for j in range(corr_matrix.shape[0]):
        plt.text(i+0.5,j+0.5,round(corr_matrix.iloc[j,i],2),ha='center',va='center')

plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.title("Correlation Heatmap")
plt.show()
```

```python
#모든 컬럼 중에서 상관관계가 있는 컬럼 5개 추출하기
import numpy as np
#spotify관련 데이터셋
s_music = music[cols]

# 상관계수 행렬
corr = s_music.corr(numeric_only=True)

# 상삼각행렬만 사용 (중복 제거)
corr_upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))

# 절대값 기준 정렬
top5 = (
    corr_upper.abs()
    .unstack()
    .dropna()
    .sort_values(ascending=False)
    .head(5)
)

print(top5)
```

Loudness(크기), Acousticness(자연 악기 사용정도) 가 높을수록 Energy가 크게 나타남  
Instrumentalness(보컬적 요소), Acousticness(자연 악기 사용정도)가 높을수록 Loudness(음량)이 큼  
Valence(긍정적)일수록 Danceability(댄스) 댄스곡일 확률이 높음  

```python
#Stream컬럼과 가장 관계있는 컬럼 3개
top3 = (
    s_music.corr(numeric_only=True)["Stream"]   # stream과의 상관계수
    .drop("Stream")                        # 자기 자신 제거
    .abs()                                 # 절대값 기준
    .sort_values(ascending=False)          # 내림차순 정렬
    .head(3)                               # 상위 3개
)

print(top3)
```

음량이 높고, 전자음의 비중이 적고 자연 악기를 많이 사용할수록, 보컬적 요소가 적을수록 스트리밍 수가 많을 것이다

넓은 범위에서 스포티파이, 유튜브 관련 행 모두를 사용해서 상관관계를 확인해보자!  
(위에서는 스포티파이 데이터만 다뤘음)

```python
# 상관계수 행렬
corr = music.corr(numeric_only=True)

# 상삼각행렬만 사용 (중복 제거)
corr_upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))

# 절대값 기준 정렬
top5 = (
    corr_upper.abs()
    .unstack()
    .dropna()
    .sort_values(ascending=False)
    .head(5)
)

print(top5)
```

스트림수가 많은 음악들이 조회수, 좋아요, 댓글 수가 많음을 알 수 있다!

지금은 수치형 데이터를 사용해서 확인했지만 문자열 컬럼, 특히 가수의 영향도 생각해 봐야함!

```python
#이 데이터셋에서 Stream수가 가장 많은 Artist는 누구일까?
top5_artist = (
    music.groupby("Artist")["Stream"]   # Artist별로 묶고
      .sum()                         # Stream 합계 계산
      .sort_values(ascending=False)  # 내림차순 정렬
      .head(10)                       # 상위 5명
)

print(top5_artist)
```
