### W1 - TASK4.3 각 칼럼(센서)와 타겟의 상관계수 계산

### W1 - TASK4.5. 이상치 확인
그래프 분석 결과, 데이터에 이상치가 상당히 존재한다는 것을 확인했다. 이상치의 존재가 제품 성공/불량에 어느정도 영향을 미치는 지 확인할 필요성이 있다.

```python
# 이상치 개수를 계산하는 함수 생성 (IQR 방식)
def count_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return ((series < lower_bound) | (series > upper_bound)).sum()

# 대상 센서 (상관계수 상위 4개)
target_sensors = ['59', '103', '510', '348']
results = []


for sensor in target_sensors:
    # 각 그룹별 데이터 추출
    pass_group = data[data['Pass/Fail'] == -1][sensor]
    fail_group = data[data['Pass/Fail'] == 1][sensor]

    # 그룹별 이상치 계산
    pass_outliers = count_outliers(pass_group)
    fail_outliers = count_outliers(fail_group)

    # 전체 대비 비율 계산
    pass_ratio = (pass_outliers / len(pass_group)) * 100
    fail_ratio = (fail_outliers / len(fail_group)) * 100

    # 결과 입력
    results.append({
        'Sensor': sensor,
        'Pass Outliers (count)': pass_outliers,
        'Pass Ratio (%)': round(pass_ratio, 2),
        'Fail Outliers (count)': fail_outliers,
        'Fail Ratio (%)': round(fail_ratio, 2)
    })

# 결과 출력
outlier_analysis = pd.DataFrame(results)
print(outlier_analysis)
```

만약 이상치 비율이 Pass 그룹에서 낮다가 Fail 그룹에서 높아졌다면, 이상치의 존재가 제품의 Fail에 큰 영향을 미치는 것이므로 이상치를 버리면 안 된다는 뜻이다. 그러나 현재는 이상치가 오히려 Pass 그룹에서 높게 관측되므로, 제품의 성공/실패 유무에 큰 영향을 끼치지 않는 것으로 판단된다. 따라서 이상치를 제거하는 것이 필요해보인다.

### W2 - TASK1.1 결측치 비율이 0.5 이상인 칼럼(센서) 제거
EDA 과정에서 데이터를 확인한 결과, 결측비율이 0.5가 넘는 센서가 다수 존재했다. 해당 센서를 제거하더라도 수 백개의 센서가 남아있기 때문에, 불필요하다고 판단되어 과감히 제거한다.

```python
# data 변수에 저장된 각 칼럼별 결측치의 비율 계산
null_ratio = data.isnull().mean()

# 결측치 비율이 0.5 이상인 칼럼들의 이름 추출
columns_to_drop = null_ratio[null_ratio >= 0.5].index.tolist()

# 기존 'data' 자료에서 해당 칼럼을 제거한 새로운 데이터를 'data_cleaned'에 저장
data_cleaned = data.drop(columns=columns_to_drop, axis=1)

# 결과 확인
print(f"기존 칼럼 수: {data.shape[1]}개")
print(f"제거 대상 칼럼(결측치 50% 이상) 수: {len(columns_to_drop)}개")
print("-" * 30)
print(f"제거 후 남은 칼럼 수: {data_cleaned.shape[1]}개")
```

### W2 - TASK1.2 값이 변하지 않는 칼럼(분산=0) 제거
앞서 기초 통계량 확인을 통해 분산이 0인 센서들을 일부 발견하였다. 해당 센서는 값이 변하지 않으므로 상수와도 같다. 불필요하기에 해당 칼럼들을 제거한다.

```python
# 각 칼럼별로 고유한 값(Unique Value)의 개수 확인
unique_counts = data_cleaned.nunique()

# Unique Value 가 1인 칼럼들의 이름 추출
single_value_columns = unique_counts[unique_counts == 1].index.tolist()

# 해당 칼럼들을 제거하여 새로운 변수인 'data_dropped'에 저장
data_dropped = data_cleaned.drop(columns=single_value_columns, axis=1)

# 결과 확인
print(f"기존 칼럼 수: {data_cleaned.shape[1]}개")
print(f"제거 대상 칼럼(값이 변하지 않는 칼럼) 수: {len(single_value_columns)}개")
print("-" * 30)
print(f"제거 후 남은 칼럼 수: {data_dropped.shape[1]}개")
```

### W2 - TASK1.3 이상치 처리하기  
앞서 분석했을 때 이상치가 결과에 큰 영향을 주지 않는 것으로 판단되어, 이상치를 처리한다. 불량 데이터의 수가 매우 적으므로 이상치를 아예 삭제하는 것이 아닌, 특정 임계값으로 값을 깎아서 고정시키는 방법인 윈저라이징(Clipping) 방법을 이용한다.

```python
# 'Time'과 'Pass/Fail' 컬럼은 제외하고 센서 데이터만 선택
features = data_dropped.drop(columns=['Time', 'Pass/Fail']).columns

# 각 센서별로 상위 1%와 하위 1%에서 Clipping 수행
for col in features:
    lower_limit = data_dropped[col].quantile(0.01)
    upper_limit = data_dropped[col].quantile(0.99)

    data_dropped[col] = data_dropped[col].clip(lower=lower_limit, upper=upper_limit)
data_final = data_dropped.copy()
# data_final 에 저장
print(f"윈저라이징 완료, 데이터 형태: {data_final.shape}")
```

앞서 W1 - TASK4.5 이상치 확인에서 수행했던 코드를 다시 실행해본다.

```python
# 이상치 개수를 계산하는 함수 생성 (IQR 방식)
def count_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return ((series < lower_bound) | (series > upper_bound)).sum()

# 대상 센서 (상관계수 상위 4개)
target_sensors = ['59', '103', '510', '348']
results = []


for sensor in target_sensors:
    # 각 그룹별 데이터 추출
    pass_group = data_final[data_final['Pass/Fail'] == -1][sensor]
    fail_group = data_final[data_final['Pass/Fail'] == 1][sensor]

    # 그룹별 이상치 계산
    pass_outliers = count_outliers(pass_group)
    fail_outliers = count_outliers(fail_group)

    # 전체 대비 비율 계산
    pass_ratio = (pass_outliers / len(pass_group)) * 100
    fail_ratio = (fail_outliers / len(fail_group)) * 100

    # 결과 입력
    results.append({
        'Sensor': sensor,
        'Pass Outliers (count)': pass_outliers,
        'Pass Ratio (%)': round(pass_ratio, 2),
        'Fail Outliers (count)': fail_outliers,
        'Fail Ratio (%)': round(fail_ratio, 2)
    })

# 결과 출력
outlier_analysis = pd.DataFrame(results)
print(outlier_analysis)
```

이상치 처리 전 결과가 다음과 같음을 고려할 때, 103번 센서의 이상치가 7개 감소했고 59번 센서의 이상치도 1개 감소했음을 알 수 있다. 이상치가 잘 처리되었다.

### W2 - TASK1.5 데이터 스케일링  
센서마다 값의 사이즈가 다르므로 스케일링을 진행해야 한다. 대표적인 스케일러는 StandardScaler, RobustScaler가 있지만 현재 데이터는 이상치가 많으므로 그에 적합한 RobustScaler를 사용한다.

```python
# 사이킷런에서 스케일러 불러오기
from sklearn.preprocessing import RobustScaler

# 객체 할당
scaler = RobustScaler()

# 특성(x)과 타겟(y) 분리하기
# 'Pass/Fail'은 우리가 맞춰야 할 타겟(y)으로 분류한다.
# 'Time' 열은 숫자 데이터가 아니라 날짜이므로 제외한다.
x = data_final_imputed.drop(columns=['Pass/Fail', 'Time'])
y = data_final_imputed['Pass/Fail']

# 데이터 학습 및 변환
x_scaled = scaler.fit_transform(x)

# 결과를 데이터프레임으로 변환 -> x_scaled_df 에 저장
x_scaled_df = pd.DataFrame(x_scaled, columns=x.columns)

# 결과 출력
print("날짜 데이터를 제외하고 데이터 스케일링 수행")
x_scaled_df.head()
```

### W2 - TASK1.6 다중공선성(Multicollinearity) 해결  
반도체 공정 특성상 여러 센서가 동일한 압력이나 온도를 측정하여 데이터가 서로 강한 상관관계를 가지는 경우가 발생한다. 중복된 정보는 모델을 복잡하게 만들고 해석을 어렵게 하므로, 서로 상관계수가 0.9 이상인 센서들 중 하나만 남기고 삭제하여 다중공선성을 해결한다.

```python
# 센서 간 상관계수 행렬 계산(절댓값)
# 현재 남아있는 센서는 446개이기 때문에, 행렬의 크기는 446 x 446 임.
corr_matrix = x_scaled_df.corr().abs()

# 중복 계산을 방지하기 위해 상삼각 행렬(Upper triangle)만 선택
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# 상관계수가 0.9 넘는 칼럼 찾기
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]

# 해당 칼럼들 제거하기 -> x_final 에 저장
x_final = x_scaled_df.drop(columns=to_drop)

print("[상관계수가 0.9 이상인 칼럼 제거]")
print(f"기존 칼럼 수: {x_scaled_df.shape[1]}개")
print(f"상관계수가 0.9 이상인 칼럼 수: {len(to_drop)}개")
print("-" * 40)
print(f"남은 칼럼 수: {len(x_final.columns)}개")
```

### W2 - TASK2 전반적 공정 변동성 변수 생성
- 현재 데이터의 칼럼명은 마스크 처리되어있어 그 실체를 알 수 없다. 이러한 상황에서 특정 변수를 조합해 새로운 의미있는 변수를 만들어 내는 것은 위험한 시도라고 생각한다. 그러나 각 행(샘플)별 통계량을 구하는 것은 해당 제품이 생산될 때 공정 전체가 얼마나 안정적이었는지를 보여주는 훌륭한 지표가 될 것이라 생각한다.

- 각 행(샘플)별로 주요 센서들의 값이 평균에서 얼마나 벗어나 있는지 그 변동의 합을 구하는 개념으로, "이 샘플의 공정에서는 전반적으로 센서들이 요동치고 있는가?" 를 수치화한 지표로 만들 것이다.

- 이를 위해 각 샘플별 센서와의 평균값(mean)과 표준편차(std) 변수를 생성한다.

```python
# 요약 변수를 생성할 센서 리스트 (Pass/Fail, Time 등 타겟 제외)
sensor_columns = [col for col in x_final.columns if col not in ['Time', 'Pass/Fail']]

# TASK 2.1 전체 센서에 대한 행별 평균값 통계량 'row_mean' 변수 생성
x_final['row_mean'] = x_final[sensor_columns].mean(axis=1)

# TASK 2.1 전체 센서에 대한 행별 표준편차 통계량 'row_std' 변수 생성
x_final['row_std'] = x_final[sensor_columns].std(axis=1)

print("새로운 변수 'row_mean', 'row_std' 생성 완료.\n")
print(x_final[['row_mean', 'row_std']])
print(f"현재 데이터 크기: {x_final.shape, y.shape}")
```

### W2 - TASK1.7 SMOTE 기법을 이용한 데이터 불균형 해소  
- 현재 데이터는 '성공'이 압도적으로 많다. 이 상태로 모델을 학습하면 모델이 무조건 '성공'이라고 예측하는 편향이 발생할 수 있다. SMOTE(Synthetic Minority Over-sampling Technique) 기법은 부족한 '실패' 데이터를 수학적으로 생성하여 성공 데이터와 실패 데이터의 비율을 맞춰주는 기법이다.   
- 현재 정상 데이터는 1463 개, 불량 데이터는 104 개이다. 불량의 비율은 약 6%이므로 이를 30%로 우선 늘려 모델을 학습하고, 성능을 더 높일 필요성이 생기면 50%를 거쳐 100%까지 늘려보는 전략을 세웠다.

```python
# SMOTE 를 위해 imbalanced-learn 라이브러리 불러오기
from imblearn.over_sampling import SMOTE

# SMOTE 객체 생성, 랜덤 시드는 42
# 초기 접근: 불량(1) 데이터를 정상(-1) 데이터의 30% 수준까지 생성해본다.
# 추후 sampling_strategy의 비율을 1까지 올려 성공과 실패의 비율을 1:1로 맞출 예정이다.
smote = SMOTE(sampling_strategy=0.3, random_state=42)

# 실패 데이터 생성
x_resampled, y_resampled = smote.fit_resample(x_final, y)

print(f"[SMOTE 전]\n실패 데이터: {len(y[y == 1])}개\n성공 데이터: {len(y[y == -1])}개\n전체 데이터 중 실패 데이터 비율: {np.round(len(y[y == 1]) / len(y) * 100, 2)}%\n")
print(f"[SMOTE 후]\n실패 데이터: {len(y_resampled[y_resampled == 1])}개\n성공 데이터: {len(y_resampled[y_resampled == -1])}개\n전체 데이터 중 실패 데이터 비율: {np.round(len(y_resampled[y_resampled == 1]) / len(y_resampled) * 100, 2)}%")
```

문제 유형 정의: target이 Pass(0)/Fail(1) 를 구분하는 것이므로 분류모델로 수행한다.

현재 샘플 수는 약 2000개, 성공 대비 실패 샘플 비율은 23%, 샘플별 특성(칼럼) 수는 253개이다. 센서의 수가 많기 때문에, 어떤 센서가 제품의 수율에 영향을 주는지 파악하는게 중요하다.

모델 선택: 센서의 수가 많기 때문에 tree 알고리즘을 선택하였다. 베이스 모델로 Random Forest를 이용해 특성 중요도 및 기타 성능을 파악한 뒤, XGBoost 모델로 최종 결정할 예정이다.

우선 데이터를 학습(train)/검증(test) 세트로 알아보기 쉽게 나눈다.

```python
# 학습: input
# 검증: target

# x_resampled(입력), y_final(출력)을 각각 input, target에 할당한다.
input = x_resampled
target = y_final
```

train_test_split 수행

```python
# train_test_split을 이용해 학습용 80%, 검증용 20%로 데이터를 나눈다. 이때 제품의 불량률이 23%로 미미하다.
# 따라서 해당 비율을 골고루 나누기 위해 startify=target 을 통해 불량률 비율을 적절히 분배한다.
from sklearn.model_selection import train_test_split

# 데이터 분할(학습용 80%, 검증용 20%)
train_input, test_input, train_target, test_target = train_test_split(input, target, test_size=0.2, random_state=42, stratify=target)
```

학습 시간을 기록하기 위해 time 라이브러리를 활용한다.

```python
# time 라이브러리 불러오기
import time
start_time = time.time()  # 시작 시간 기록

# 랜덤 포레스트 모델 생성 및 학습
from sklearn.ensemble import RandomForestClassifier

# 실패 23%인 불균형 데이터이므로. class_weight='balanced'로 두어 학습 시 가중치를 조절한다.
rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf.fit(train_input, train_target) # 모델 학습

end_time = time.time() # 종료 시간 기록
fit_time = end_time - start_time # 학습 시간 계산

print(f"학습 소요 시간: {round(fit_time, 1)}초")

# 검증 세트를 이용해 예측 데이터 생성
test_pred_rf = rf.predict(test_input)
```

불량률이 23%로 낮은 데이터이므로. 하이퍼파리미터인 class_weight를 'balanced'로 둔다. 이는 특정 클래스의 양이 너무 적을 때, 모델이 쪽수에 밀려 소수 클래스를 무시하지 않도록 가중치를 자동으로 조절해준다.

```python
# 성능 평가
from sklearn.metrics import classification_report, confusion_matrix

print("RandomForest 성능 평가\n")
print(f"학습 소요 시간: {round(fit_time, 1)}초")
print(confusion_matrix(test_target, test_pred_rf))
print(classification_report(test_target, test_pred_rf))
```

[성능 평가 분석]
- 0이 정상, 1이 불량이다.
- [293, 0]: 실제 0인데 0으로 맞춘게 293개 / 실제 0인데 1로 잘못 예측한게 0개
- [17, 71]: 실제 1인데 0으로 놓친게 17개 / 실제 1인데 1로 맞춘게 71개

- precision: 1(불량)은 1.0으로 모델이 정상을 불량으로 오인하는 일이 없었다.

- recall: 0(정상)은 1.0으로 정상인 샘플은 모두 잘 잡고 있으나, 1(불량)에 대해서는 0.81로 불량 샘플 중 19%는 잡지 못하는 모습을 보인다.

- f1-score: 0(정상)에 대해서는 0.97으로 높은 점수를 보이나, 1(불량)에 대해서는 recall의 낮은 점수로 인해 0.89로 다소 떨어진 점수를 보인다.

- accuracy: 0.96로 높은 점수를 보이나, 정상 데이터의 비율이 높은 데이터임과 불량 데이터를 놓쳐 recall이 0.81에 달하는 것을 감안하여 수치만 높은 것으로 해석할 수 있다.

해결해야 할 문제: recall 수치가 0.81인 것은 불량 데이터 중 성공이라고 외친 비율이 19%나 된다는 뜻이다. 놓친 데이터를 더 잡아야 한다.

### (2) XGBoost

랜덤 포레스트 모델 학습을 통해 주요 센서 파악 및 개선할 점을 확인했다. 이번엔 XGBoost 모델의 성능 및 주요 센서를 파악해보겠다.

```python
from xgboost import XGBClassifier

# XGBoost 모델 생성
xgb = XGBClassifier(
    n_estimators=100,       # 나무의 수. 적절한 기준점인 100을 사용.
    learning_rate=0.1,      # 각 단계의 오차를 얼마나 반영할지 결정. 널리 쓰이는 0.1을 사용
    max_depth=4,            # 샘플 수(2000개)에 비해 변수(253개)가 상당히 많은 고차원 데이터이므로, 과적합 방지를 위해 깊이를 4로 제한
    scale_pos_weight=3.34,  # scale_pos_weight 파라미터는 데이터 불균형 문제를 해결하기 위한 파라미터이다.
                            # 정상 77%, 불량 23%의 데이터이므로 (정상 샘플 수 / 불량 샘플 수) = 3.34를 대입한다.
                            # 불량 데이터를 학습할 때 발생하는 오차에 약 3.3배의 가중치를 두어, 결과적으로 Recall을 높이는 효과가 있다.
    eval_metric='logloss',  # 현재와 같은 이진 분류 문제에서 표준적으로 사용되는 손실 함수이다.
    random_state=42
)

# 모델 학습
start_time = time.time() # 시작 시간 기록

xgb.fit(train_input, train_target)

end_time = time.time() # 종료 시간 측정
fit_time = end_time - start_time # 학습 시간 계산
print(f"학습 소요 시간: {round(fit_time, 1)}초")

# 검증 세트를 이용해 예측 데이터 생성
test_pred_xgb = xgb.predict(test_input)
```

XGBoost 성능 평가

```python
# 성능 평가
print("XGBoost 성능 평가\n")
print(f"학습 소요 시간: {round(fit_time, 1)}초")
print(confusion_matrix(test_target, test_pred_xgb))
print(classification_report(test_target, test_pred_xgb))
```

### (3) XGBoost Cross Validation

데이터 샘플 수가 2000개로 적은 상황이므로 교차 검증이 필요하다고 느꼈다. 이전에 만든 XGBoost 모델에 교차 검증을 시도한다.

우선 클래스의 불균형(23%)이 있으므로 StratifiedKFold 를 사용하여 각 폴드(Fold)마다 실패 데이터의 비율을 일정하게 유지하는 것이 필요하다.

```python
from sklearn.model_selection import cross_validate, StratifiedKFold

start_time = time.time()  # 시작 시간 기록



# 교차 검증 설정(5-Fold)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 측정할 지표 설정(precision, recall, f1-score, accuracy)
scores = ['precision', 'recall', 'f1', 'accuracy']

# 교차 검증 실행
cv_results = cross_validate(xgb, input, target, cv=skf, scoring=scores)



end_time = time.time() # 종료 시간 기록
fit_time = end_time - start_time # 학습 시간 계산
print(f"학습 소요 시간: {round(fit_time, 1)}초\n")

# 결과 시각화 및 출력
print("불량(1)에 대한 XGBoost 5-Fold Cross Validation 결과")


metrics = ['test_precision', 'test_recall', 'test_f1', 'test_accuracy']

for metric in metrics:
    mean_score = np.mean(cv_results[metric])
    std_score = np.std(cv_results[metric])
    print(f"{metric[5:].capitalize():10}: 평균 {mean_score:.4f} / 표준편차 {std_score:.4f}")

```
